# **Convolutional Neural Network**
A Convolutional Neural Network, also known as CNN or ConvNet, is a class of neural networks that specializes in processing data that has a grid-like topology, such as an image.Each neuron works in its own receptive field and is connected to other neurons in a way that they cover the entire visual field.

## **Normalizing the pixel of the image**

```python
normalize_layer = keras.layers.experimental.preprocessing.Rescaling(1./255)
```

```python
normalize_ds = train_ds.map(lambda x, y: (normalize_layer(x), y))
```

## **Data Augmentation**
- horizonal flip - it will randomly flip the image horizontally
- rotation range - it will rotate image randomly
- zoom range - it will zoom the image
- shear range - Shear means that the image will be distorted along an axis, mostly to create or rectify the perception angles 

```python
data_augmentation = keras.Sequential([
    keras.layers.experimental.preprocessing.RandomFlip('horizontal', input_shape=(128, 128, 3)),
    keras.layers.experimental.preprocessing.RandomZoom(0.1),
    keras.layers.experimental.preprocessing.RandomRotation(0.1)
])
```

## **CNN Model**
- In a CNN, each layer has two kinds of parameters : weights and biases. The total number of parameters is just the sum of all weights and biases. = Number of weights of the Conv Layer. ... = Size (width) of kernels used in the Conv Layer
- `padding` same means if you use a stride of 1, the layer's outputs will have the same spatial dimensions as its inputs
- `ReLU` is an activation function
- `Max Pool Layer` - Max pooling is a pooling operation that selects the maximum element from the region of the feature map covered by the filter. Thus, the output after max-pooling layer would be a feature map containing the most prominent features of the previous feature map.

```python
model = keras.models.Sequential([
    data_augmentation,
    ''' normalizting the pixel of the images '''
    keras.layers.experimental.preprocessing.Rescaling(1./255, input_shape=(128, 128, 3)),
    
    ''' 1st CNN Layer '''
    keras.layers.Conv2D(32, 3, padding='same', activation='relu'),
    keras.layers.MaxPool2D(),
    
    ''' 2nd CNN Layer '''
    keras.layers.Conv2D(64, 3, padding='same', activation='relu'),
    keras.layers.MaxPool2D(),
    
    ''' 3rd CNN Layer '''
    keras.layers.Conv2D(128, 3, padding='same', activation='relu'),
    keras.layers.MaxPool2D(),
    keras.layers.Dropout(0.2), 
    
    ''' flattening the output '''
    keras.layers.Flatten(),
    
    ''' Hidden Layer '''
    keras.layers.Dense(256),
    ''' Classification Layer'''
    keras.layers.Dense(num_classes)
])
model.summary()
```

```python
''' initializing the loss, optimizer and metrics '''
loss = keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optimizer = keras.optimizers.Adam(learning_rate=0.001)
metrics = ['accuracy']

model.compile(loss=loss,
              optimizer=optimizer,
              metrics=metrics)
```

```python
''' training of model '''
epochs = 15
callbacks = keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    verbose=2
)

history = model.fit(train_ds, validation_data=val_ds, epochs=epochs, callbacks=[callbacks], verbose=2)
```

```python
''' plotting training loss and validation loss graph '''
val_loss = history.history['val_loss']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']

plt.figure(figsize=(10, 10))
plt.plot(val_loss, 'r-o', label='validation_loss')
plt.plot(train_loss, 'b-o', label='train_loss')
plt.xlabel('epochs')
plt.ylabel('Loss')
plt.legend('Loss vs No. of epochs')
plt.plot()
```

```python
''' plotting training and validation accuracy graph '''
train_acc = history.history['accuracy']

plt.figure(figsize=(10, 10))
plt.plot(val_acc, 'r-o', label='val_accuracy')
plt.plot(train_acc, 'b-o', label='train_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epochs')
plt.show()
```
